```python
def pre_match_tweet(batting_team_id, bowling_team_id,venue_id=None):
    #number of matches with opposition
    h2h=0
    total_runs_scored=0
    for matches in match_insight_obj[batting_team_id].values():
        if matches["opposition"]==bowling_team_id:
            h2h+=1
            total_runs_scored += matches["batting_insights"]["runs_scored"]["Total"]
    return total_runs_scored, h2h        
            
```
